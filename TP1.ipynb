## 2 - Des boucles pour répéter des instructions :

Il peut être répétitif d’écrire des lignes de code. On peut utiliser une boucle pour répéter des instructions.

<div class="alert alert-success" role="alert">

    une boucle bornée s'écrit avec le mot clé `for` :
    
```python
for i in range(...):        # <-- ici le compteur est i mais ça pourrait être un autre nom
    instructions            # <-- l'indentation (le décalage) est obligatoire
```
- **Il ne faut pas oublier les  `:`**  à la fin de la première ligne.
- L'instruction `range()` renvoie une suite d'entiers :
    - `range(n)` : permet de répéter `n` fois (entiers de 0 inclus à n exclu)
    - `range(n,m)` : entiers de n inclus à m exclu
</div>

Les fonctions permettent d’organiser le code principal et d’éviter les répétitions.

 On lui donne un nom qui servira à l’utiliser à chaque fois qu’on le souhaite.
