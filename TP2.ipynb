## 2 - Des rectangles de couleur :

On commence par créer une fonction rectangle qui dessine un rectangle vertical

<div class="alert alert-info" role="alert">
  <strong>Travail à faire sur le notebook : </strong> <br>
    Pour la cellule de code suivante, appuyez sur <strong>"shift + entrée"</strong> pour lancer le code Python contenu dans chaque cellule
</div>

````python
# ne pas MODIFIER le code suivant
def rectangle(cote):
    forward(cote)
    left(90)
    forward(1.5*cote)
    left(90)
    forward(cote)
    left(90)
    forward(1.5*cote)
    left(90)
````

````python
rectangle(100)
done()
````

Le code suivant indique comment tracer un rectangle coloré :

````python
speed(5) # pour modifier la vitesse de la tortue

color("#FF00FF") # couleur RVB en hexadécimal
# à partir d'ici, on peut modifier le code

begin_fill() # indique qu'il faut remplir la forme dessinée

rectangle (100)

end_fill() # pour terminer le remplissage
up()

done()
````

On veut dessiner le drapeau français.  
On a tracé un rectangle mais il n'est pas de la bonne couleur


<div class="alert alert-success" role="alert">

Les couleurs sont définies par trois nombres en hexadécimal ( de 0 à F) pour les 3 couleurs primaires (rouge, vert, bleu) RVB.   
Chaque nombre peut varier de 00 (cette couleur n’est pas présente) à FF (cette couleur est au maximum).

L’instruction `color( ????)` permet ainsi de choisir la couleur de dessin de la tortue.

Exemples :  
 ```color("blue")``` correspondra à la couleur bleu primaire

```color("#FF00FF")``` correspondra à la superposition de la couleur rouge la plus intense et d’un bleu le plus intense : on obtient un magenta
  
    
```color("#00FF00")``` correspondra à 0 pour le rouge et le bleu mais 100% pour le vert

On définit une fonction qui dessine un arc de cercle 

````python
def arc(rayon, couleur):
    pensize(10)
    up()
    setheading(90)
    down()
    color(couleur)
    circle(rayon, 180)
    up()
    circle(rayon, -180)
````

````python
arc(100, "blue")
done()
````
