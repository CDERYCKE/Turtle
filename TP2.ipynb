```````python
speed(5)
shape("turtle") 

# fonction qui définit le tracé d'un carré valeur du coté à mettre entre parenthèses
def carre (cote) : # ne pas modifier
  for i in range(4):
    forward(cote)
    left(90)

up() # pour lever le crayon et ne pas écrire
goto # pour déplacer la tortue
down() # pour baisser le crayon et écrire

# on peut modifier la suite

for i in range(2):
  carre(5)
  up()
  forward(20)
  down()

done() # pour indiquer que le dessin est terminé
```````

## 2 - Des rectangles de couleur :

On commence par créer une fonction rectangle qui dessine un rectangle vertical

```````python
# ne pas MODIFIER le code suivant
def rectangle(cote):
    forward(cote)
    left(90)
    forward(1.5*cote)
    left(90)
    forward(cote)
    left(90)
    forward(1.5*cote)
    left(90)
```````

```````python
rectangle(100)
done()
```````

```````python
colormode(255) # les couleurs sont choisis en mélangeant Rouge, Vert et Bleu : mini = 0 , maxi = 255
speed(5) # pour modifier la vitesse de la tortue


# à partir d'ici, on peut modifier le code

color (0,255,0) # pour choisir la teinte verte au maximum et ne pas choisir ni bleu ni rouge
begin_fill() # indique qu'il faut remplir la forme dessinée

rectangle (100)

end_fill() # pour terminer le remplissage
up()

# on a tracé un rectangle et il n'est pas de la bonne couleur
# à vous de modifier le code ci-dessus
# puis avec des copier -coller de faire dessiner aussi les 2 autres bandes du drapeau
done()
```````

<div class="alert alert-success" role="alert">

Les couleurs sont définies par un triplet de nombres (rouge, vert, bleu) RVB.  Chaque nombre peut varier de 0 (cette couleur n’est pas présente) à 255 (cette couleur est au maximum).
L’instruction ```color(r,g,b)`````` permet ainsi de choisir la couleur de dessin de la tortue.

Exemples :  
 ```color(255,0,0)``` correspondra à la couleur rouge la plus intense

```color(255,0,200)``` correspondra à la superposition de la couleur rouge la plus intense et d’un bleu à mi intensité
